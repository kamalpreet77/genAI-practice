Opening file with an URL

```python
import requests
from io import StringIO
url =""
headers={}
req=requests.get(url,headers=headers)
data=StringIO(req.text)
pd.read_csv(data)
```

Sep Parameter

```python
pd.read_csv('placement.csv',sep='\t', names=['col1','col2'])
# tsv file and names the columns
```

Index_col Parameter

```python
pd.read_csv('placement.csv',index_col='ssc_p')
```

Header Parameter

```python
pd.read_csv('placement.csv',header=1)
# if you have header in the first row
```

use_col Parameter

```python
pd.read_csv('placement.csv',usecols=['ssc_p','hsc_p'])
```

Squeeze Parameter

```python
pd.read_csv('placement.csv',squeeze=True)
```

skiprows/nrows Parameter

```python
pd.read_csv('placement.csv',skiprows=[0,1])
pd.read_csv('placement.csv',nrows=10)

```

Encoding Parameter

```python
pd.read_csv('placement.csv',encoding='latin1')
```

Skip Bad Lines

```python
pd.read_csv('placement.csv',error_bad_lines=False)
```

Dtypes parameter

```python
pd.read_csv('placement.csv',dteyp={'ssc_p':float,'hsc_p':float})
```

Handling Dates

```python
pd.read_csv('placement.csv',parse_dates=['ssc_p','hsc_p'])
```

Loading a huge dataset in chunks

```python
dfs=pd.read_csv('placement.csv',chunksize=10)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
